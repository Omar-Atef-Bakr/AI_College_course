```python
import pygame
import threading
import time

# Define puzzle board dimensions and colors
WINDOW_SIZE = 300
GRID_SIZE = 3
GRID_WIDTH = WINDOW_SIZE // GRID_SIZE
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

current_puzzle = None
puzzle_history = []
current_puzzle_index = -1
screen = None

def draw_grid(screen):
    for i in range(1, GRID_SIZE):
        pygame.draw.line(screen, BLACK, (i * GRID_WIDTH, 0), (i * GRID_WIDTH, WINDOW_SIZE))
        pygame.draw.line(screen, BLACK, (0, i * GRID_WIDTH), (WINDOW_SIZE, i * GRID_WIDTH))

def draw_puzzle(screen, puzzle):
    font = pygame.font.Font(None, 36)
    for row in range(GRID_SIZE):
        for col in range(GRID_SIZE):
            cell_value = puzzle[row][col]
            if cell_value != 0:
                cell_text = font.render(str(cell_value), True, BLACK)
                cell_rect = cell_text.get_rect(center=(col * GRID_WIDTH + GRID_WIDTH // 2, row * GRID_WIDTH + GRID_WIDTH // 2))
                screen.blit(cell_text, cell_rect)

def visualize_puzzle(puzzle):
    global current_puzzle
    current_puzzle = puzzle
    global current_puzzle_index
    current_puzzle_index += 1
    puzzle_history.append(puzzle)

def next_puzzle():
    global current_puzzle_index
    current_puzzle_index = min(current_puzzle_index + 1, len(puzzle_history) - 1)
    return puzzle_history[current_puzzle_index]

def prev_puzzle():
    global current_puzzle_index
    current_puzzle_index = max(current_puzzle_index - 1, 0)
    return puzzle_history[current_puzzle_index]

def puzzle_thread():
    pygame.init()
    global screen
    screen = pygame.display.set_mode((WINDOW_SIZE, WINDOW_SIZE))
    pygame.display.set_caption('8-Puzzle Visualization')

    while True:
        time.sleep(0.1)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                return

        if current_puzzle is not None:
            screen.fill(WHITE)
            draw_grid(screen)
            draw_puzzle(screen, current_puzzle)
            pygame.display.flip()

def start_puzzle_thread():
    puzzle_thread_thread = threading.Thread(target=puzzle_thread)
    puzzle_thread_thread.daemon = True
    puzzle_thread_thread.start()

# Example usage:
start_puzzle_thread()
time.sleep(2)  # Wait for the thread to start (adjust as needed)

puzzle1 = [
    [1, 2, 3],
    [8, 0, 4],
    [7, 6, 5]
]

visualize_puzzle(puzzle1)

# Function to update the display based on the current state
def update_display():
    global current_puzzle
    current_puzzle = puzzle_history[current_puzzle_index]

# Create "Next" and "Previous" buttons
next_button = pygame.Rect(10, 10, 80, 30)
prev_button = pygame.Rect(100, 10, 80, 30)


while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.MOUSEBUTTONDOWN:
            mouse_pos = event.pos
            if next_button.collidepoint(mouse_pos):
                current_puzzle = next_puzzle()
                update_display()
            if prev_button.collidepoint(mouse_pos):
                current_puzzle = prev_puzzle()
                update_display()

    # Draw the buttons
    screen.fill(WHITE)
    draw_grid(screen)
    draw_puzzle(screen, current_puzzle)
    pygame.draw.rect(screen, (0, 128, 255), next_button)
    pygame.draw.rect(screen, (0, 128, 255), prev_button)

    # Add button text
    font = pygame.font.Font(None, 36)
    next_text = font.render("Next", True, BLACK)
    prev_text = font.render("Prev", True, BLACK)
    screen.blit(next_text, (25, 15))
    screen.blit(prev_text, (115, 15))

    pygame.display.flip()
```
